```python
df = pd.read_csv('../../data/processed/complaints-after-state-updates.csv')
```

<h3>How many facility ids are associated with multiple online names?</h3>

```python
id_name = df[['facility_id','fac_name']][df['fac_name'].notnull()].drop_duplicates().rename(columns={'fac_name':'online_name'})
```

```python
id_name.head()
```

```python
names_per_id = id_name.groupby('facility_id').count().reset_index()
```

```python
names_per_id.head()
```

```python
names_per_id[names_per_id['online_name']>1].count()[0]
```
